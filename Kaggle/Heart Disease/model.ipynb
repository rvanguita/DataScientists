```python
def cutData(data, column, porcentage, count = True):
    if count == True:
        a = data[data[column] < data[column].quantile(porcentage)]
    else:
        a = data[data[column] > data[column].quantile(porcentage)]
    return a
```

```python
def reportML(dataFrame, algoritmo, opt = 2):
    X = dataFrame.drop('target', axis = 1)
    y = dataFrame['target']
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state= 42)
    
    if opt == 0:
        ml = algoritmo(max_iter = 1000)
    elif opt == 1:
        ml = algoritmo(n_estimator = 1000)
    elif opt == 2:
        ml = algoritmo()
    
    
    ml.fit(X_train, y_train)
    score_train = ml.score(X_train, y_train)
    print(f"Training Accuracy: {score_train * 100:.2f}%")
    
    score_test = ml.score(X_test, y_test)
    print(f"Test Accuracy: {score_test * 100:.2f}%")
    
    
    y_foreseen = log.predict(X_test)
    
    accuracy_score(y_test, y_foreseen)
    
    print(f"--- Classification Report ---")
    print(classification_report(y_test, y_foreseen))
    
    print(f"--- Confusion Matrix ---")    
    confusion = confusion_matrix(y_test, y_foreseen)
    sns.heatmap(confusion, annot = True, cmap = 'Blues')
    
    
    return score_train, score_test
```

```python
log_train, log_test = reportML(dataHeart_quantile, LogisticRegression, opt = 0)
```

```python
tree_train, tree_test = reportML(dataHeart_quantile, DecisionTreeClassifier)
```

```python
ada_train, ada_test = reportML(dataHeart_quantile, AdaBoostClassifier)
```

```python
kn_train, kn_test = reportML(dataHeart_quantile, KNeighborsClassifier)
```

```python
rand_train, rand_test = reportML(dataHeart_quantile, RandomForestClassifier)
```
