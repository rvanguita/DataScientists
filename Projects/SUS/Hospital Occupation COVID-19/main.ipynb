```python
def cutData(dataFrame, column, percentage, count = True):
    if count == True:
        a = dataFrame[dataFrame[column] < dataFrame[column].quantile(percentage)]
    else:
        a = dataFrame[dataFrame[column] > dataFrame[column].quantile(percentage)]
    return a
```
